```python
# importing libraries
import pandas as pd
import glob
import numpy as np
import os
import io
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Put 2018.csv data into dataframe to manipulate
df_2018 = pd.read_csv('/content/drive/MyDrive/Raw Data/2018.csv')
```

```python
#Print out dataframe
df_2018
```

```python
#Dropping all duplicate rows
df_2018.drop_duplicates(inplace = True)
```

```python
#Print out dataframe
#No duplicates were dropped 7213446 rows before and 7213446 rows rows present after the drop
df_2018
```

```python
# Get indexes where CANCELLATION_CODE column has value A - Carrier Caused Cancellation
indexNames = df_2018[df_2018['CANCELLATION_CODE'] == 'A'].index
 
# Delete these row indexes from dataFrame
df_2018.drop(indexNames , inplace=True)
```

```python
#Print out dataframe
df_2018
```

```python
# Get indexes where CANCELLATION_CODE column has value B - Weather Cancellation
indexNames = df_2018[df_2018['CANCELLATION_CODE'] == 'B'].index
 
# Delete these row indexes from dataFrame
df_2018.drop(indexNames , inplace=True)
```

```python
#Print out dataframe
df_2018
```

```python
# Get indexes where CANCELLATION_CODE column has value C - National Aviation System Cancellation
indexNames = df_2018[df_2018['CANCELLATION_CODE'] == 'C'].index
 
# Delete these row indexes from dataFrame
df_2018.drop(indexNames , inplace=True)
```

```python
#Print out dataframe
df_2018
```

```python
# Get indexes where CANCELLATION_CODE column has value D - Security Cancellation
indexNames = df_2018[df_2018['CANCELLATION_CODE'] == 'D'].index
 
# Delete these row indexes from dataFrame
df_2018.drop(indexNames , inplace=True)
```

```python
#Print out dataframe
df_2018
```

```python
#Making a note that there were zero duplicate rows identified and removed
#We started with 7213446 rows and ended with 7096862 rows after removing cancelled flights
#116,584 rows of cancelled flight data were removed from the dataframe
```

```python
#Check for flights which have a departure time less than 15 minutes delayed
#We are defining a "late" flight as having a departure delay of 15 minutes or more

df_2018[df_2018.DEP_DELAY < 15 ].count()
```

```python
# Filter out flights with departure delays of 15 minutes or more
df_2018_filtered = df_2018[(df_2018['DEP_DELAY'] >= 15)]
```
